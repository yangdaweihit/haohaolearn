# [Using ggplot in Python: Visualizing Data With plotnine](https://realpython.com/ggplot-python/)

```python
from plotnine.data import economics, mpg, huron
from plotnine import ggplot, aes, facet_grid, geom_line, geom_point, stat_bin, geom_bar, geom_boxplot, scale_x_timedelta, labs, coord_flip
from plotnine import theme_dark, theme_xkcd
```

```python
economics
```

```python
(
    ggplot(economics)  # What data to use
    + aes(x="date", y="pop")  # What variable to use
    + geom_line()  # Geometric object to use for drawing
)
```

```python
ggplot(economics) + aes(x="date", y= "unemploy") + geom_line()
```

```python
mpg
```

```python
ggplot(mpg) + aes(x="class", y= "hwy") + geom_point()
```

```python
ggplot(mpg) + aes(x="class") + geom_bar()
```

```python
huron
```

```python
ggplot(huron) + aes(x="level") + stat_bin(bins=30) + geom_bar()
```

```python
(
  ggplot(huron)
  + aes(x="factor(decade)", y="level")
  + geom_boxplot()
)
```

```python
(
    ggplot(economics)
    + aes(x="date", y="pop")
    + scale_x_timedelta(name="Years since 1970")
    + labs(title="Population Evolution", y="Population")
    + geom_line()
)
```

```python
ggplot(mpg) + aes(x="class") + geom_bar() + coord_flip()
```

```python

(
    ggplot(mpg)
    + facet_grid(facets="year~class")
    + aes(x="displ", y="hwy")
    + labs(
        x="Engine Size",
        y="Miles per Gallon",
        title="Miles per Gallon for Each Year and Vehicle Class",
    )
    + geom_point()
)
```

```python
(
    ggplot(mpg)
    + facet_grid(facets="year~class")
    + aes(x="displ", y="hwy")
    + labs(
        x="Engine Size",
        y="Miles per Gallon",
        title="Miles per Gallon for Each Year and Vehicle Class",
    )
    + geom_point()
    + theme_dark()
)
```

```python
(
    ggplot(mpg)
    + facet_grid(facets="year~class")
    + aes(x="displ", y="hwy")
    + labs(
        x="Engine Size",
        y="Miles per Gallon",
        title="Miles per Gallon for Each Year and Vehicle Class",
    )
    + geom_point()
    + theme_xkcd()
)
```

```python
(
    ggplot(mpg)
    + aes(x="cyl", y="hwy", color="class")
    + labs(
        x="Engine Cylinders",
        y="Miles per Gallon",
        color="Vehicle Class",
        title="Miles per Gallon for Engine Cylinders and Vehicle Classes",
    )
    + geom_point()
)
```
