# [Artist tutorial](https://matplotlib.org/stable/tutorials/intermediate/artists.html#sphx-glr-tutorials-intermediate-artists-py)

Matplotlib API分三个层：
1. `matplotlib.backend_bases.FigureCanvas`是一块区域用来画图；
2. `matplotlib.backend_bases.Renderer`是一个用来在`FigureCanvas`上画图的方法对象；
3. `matplotlib.artist.Artist`是一类在canvas上画图的对象。它是用户95%时间使用的对象。

`Artist`有两个类型：

- 图元: Line2D, Rectangle, Text, AxesImage等；
- 容器：Figure, Axes，Axis等。


生成`Figure`实例的标准做法是：

- 用`Figure`生成一个或多个`Axes`或`subplot`实例；
- 并用`Axes`实例帮肋方法生成图元。

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(2, 1, 1) # two rows, one column, first plot
```

- `Axes`是Matplotlib API中最重要的类。它是一块大多数对象存在的绘图区域，有很多方法：plot(), text(), hist(), imshow()生成最常见的图元：Line2D, Text, Rectangle, AxesImage等。
- 这些方法在获取数据(numpy数组和字符串)生成`Artist`实例，将其放在容器中，并绘制出来。
- 很多人熟悉`Subplot`，它生成一个矩形行列格形式的`Subplot`实例。
- 若要在任意位置生成一个`Axes`，则使用`add_axes()`方法，它接受一个列表`[left, bottom, width, height]`，所有值都在[0,1]之间。

```python
fig2 = plt.figure()
ax2 = fig2.add_axes([0.15, 0.1, 0.7, 0.3])
ax3 = fig2.add_axes([0.4, 0.3, 0.3 ,0.3])
```

```python
import numpy as np
t = np.arange(0.0, 1.0, 0.01)
s = np.sin(2*np.pi*t)
line, = ax.plot(t, s, color='blue', lw=2)  # ax是前面fig.add_subplot产生的Axes实例
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
fig.subplots_adjust(top=0.8)
ax1 = fig.add_subplot(211)
ax1.set_ylabel('volts')
ax1.set_title('a sine wave')

t = np.arange(0.0, 1.0, 0.01)
s = np.sin(2*np.pi*t)
line, = ax1.plot(t, s, color='blue', lw=2)

# Fixing random state for reproducibility
np.random.seed(19680801)

ax2 = fig.add_axes([0.15, 0.1, 0.7, 0.3])
n, bins, patches = ax2.hist(np.random.randn(1000), 50,
                            facecolor='green', edgecolor='red')
ax2.set_xlabel('time (s)')

plt.show()
```

```python
import matplotlib.lines as lines

fig = plt.figure()

l1 = lines.Line2D([0, 1], [0, 1], transform=fig.transFigure, figure=fig)
l2 = lines.Line2D([0, 1], [1, 0], transform=fig.transFigure, figure=fig)
fig.lines.extend([l1, l2])

plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)
ax1.annotate('I am ax1', (0.5,0.5),xycoords='axes fraction', va='center',ha='center')
ax2.annotate('I am ax2', (0.5,0.5),xycoords='axes fraction')
plt.tight_layout()
```
